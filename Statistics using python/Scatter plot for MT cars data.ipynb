```python
!pip install seaborn
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('D:\\Data science with Python\\Assignments\\MT Cars\\auto_data.csv')
```

```python
df.head()
```

```python
#plot mgp ,weight and origin rel using seaborn pairplot
```

```python
sns.set_style('darkgrid')
sns.pairplot(df,
             y_vars=['mpg'],
             x_vars=['weight'],
             hue='origin',
             kind='reg',
             height=6
            )
```

```python
#remove shaded CI region
```

```python
sns.set_style('darkgrid')
sns.pairplot(df,
             plot_kws={'ci': None},
             y_vars=['mpg'],
             x_vars=['weight'],
             hue='origin',
             kind='reg',
             height=6
            )
```
