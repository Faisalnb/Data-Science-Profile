```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
labels='Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza & Pneumonia', 'Sepsis','Others'
sizes=[62, 5, 4, 2, 1 , 26]
```

```python
 # include plot area
plt.figure(figsize=(10,10))
#enter givn data
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=70)
#set equal axeses
plt.axis('equal')
#enter title for plot
plt.title('Cause of Deaths,Ohio :2012')
#show
plt.show()
```
