```python
real_estate_df.drop(['BLOCKID','SUMLEVEL'],axis=1,inplace=True)
```

```python
real_estate_df.isna().sum()
```

```python
# Imputing missing values with median
missing_train_cols=[]
for col in real_estate_df.columns:
    if real_estate_df[col].isna().sum() !=0:
        missing_train_cols.append(col)
```

```python
# Missing cols are all numerical variables
for col in real_estate_df.columns:
    if col in (missing_train_cols):
        real_estate_df[col].replace(np.nan, real_estate_df[col].median(),inplace=True)
```

```python
test_df=pd.read_csv("D:\PG_DSBA_Business_Analytics_\Data science Capstone\Project 1\\test.csv")
test_df.drop(['BLOCKID','SUMLEVEL'],axis=1,inplace=True)
```

```python
# Missing cols are all numerical variables
for col in test_df.columns:
    if col in (missing_train_cols):
        test_df[col].replace(np.nan, test_df[col].median(),inplace=True)
```

```python
# Build a linear Regression model to predict the total monthly expenditure for home mortgages loan. 
# Please refer â€˜deplotment_RE.xlsxâ€™. Column hc_mortgage_mean is predicted variable. This is the mean monthly mortgage and owner costs of specified geographical location. 
# Note: Exclude loans from prediction model which have NaN (Not a Number) values for hc_mortgage_mean.
```

```python
real_estate_df["type"].replace('City',1,inplace=True)
real_estate_df["type"].replace('Urban',2,inplace=True)
real_estate_df["type"].replace('Town',3,inplace=True)
real_estate_df["type"].replace('CDP',4,inplace=True)
real_estate_df["type"].replace('Village',5,inplace=True)
real_estate_df["type"].replace('Borough',6,inplace=True)
```

```python
test_df["type"].replace('City',1,inplace=True)
test_df["type"].replace('Urban',2,inplace=True)
test_df["type"].replace('Town',3,inplace=True)
test_df["type"].replace('CDP',4,inplace=True)
test_df["type"].replace('Village',5,inplace=True)
test_df["type"].replace('Borough',6,inplace=True)
```

```python
feature_cols=['COUNTYID','STATEID','zip_code','type','pop', 'family_mean',
'second_mortgage', 'home_equity', 'debt','hs_degree',
'pct_own', 'married','separated', 'divorced']
```

```python
x_train=real_estate_df[feature_cols]
y_train=real_estate_df['hc_mortgage_mean']
```

```python
x_test=test_df[feature_cols]
y_test=test_df['hc_mortgage_mean']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error,accuracy_score
```

```python
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.fit_transform(x_test)
```

```python
linereg=LinearRegression()
linereg.fit(x_train_scaled,y_train)
```

```python
y_pred=linereg.predict(x_test_scaled)
```

```python
y_pred.shape
```

```python
print("Overall R2 score of linear regression model", r2_score(y_test,y_pred))
print("Overall RMSE of linear regression model", np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# The Accuracy and R2 score are good
```
