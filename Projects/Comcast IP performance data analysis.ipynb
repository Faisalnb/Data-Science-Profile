```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
d_parser=lambda x:pd.datetime.strptime(x,"%d-%m-%y")
comcast_df=pd.read_csv('G:\My Drive\Assignments\Assessment_3\\Comcast_telecom_complaints_data.csv',
                       parse_dates=['Date'],date_parser=d_parser)
```

```python
comcast_df
```

```python
comcast_df=comcast_df.set_index(['Date'])
```

```python
comcast_df.sort_values(by='Date')
```

```python
comcast_df.resample('M').agg({"Customer Complaint":'count'}).rename(columns={'Customer Complaint':'Monthly_Complaints'})
```

```python
comcast_df.resample('D').agg({"Customer Complaint":'count'}).rename(columns={'Customer Complaint':'Daily_Complaints'})
```

```python
monthly_complaints,daily_complaints=comcast_df.resample('M').agg({"Customer Complaint":'count'}).rename(columns={'Customer Complaint':'Monthly_Complaints'}),
                                    comcast_df.resample('D').agg({"Customer Complaint":'count'}).rename(columns={'Customer Complaint':'Daily_Complaints'})
```

```python
ax=monthly_complaints.plot(kind='line',figsize=(15,5))
daily_complaints.plot(kind='line',ax=ax,figsize=(15,5))
```

```python
daily_complaints.mean()
```

```python
comcast_df.replace('Closed','Close',inplace=True)
comcast_df.replace('Solved','Close',inplace=True)
```

```python
comcast_df.replace('Open','Open',inplace=True)
comcast_df.replace('Pending','Open',inplace=True)
```

```python
comcast_df.head()
```

```python
state_grp=comcast_df.groupby(['State'])
```

```python
state_grp
```

```python
data=state_grp['Status'].value_counts().to_frame();data
```

```python
data1=data.unstack(level=1).fillna(0);data1.head()
```

```python
# plotting stacked Horizontal Bar Chart with all the columns
plt.figure(figsize=(20,30))
data1.plot.barh(stacked=True, title='Summary of Comcast complaints in the year 2015', color=("green", "red"),figsize=(20,30))
plt.show()
```
