```
from pyspark.sql.types import LongType,StructField,StructType
```

```
def dfZipWithIndex (df, offset=1, colName='rowId'):
    new_schema= StructType(
    [StructField(colName,LongType(),True)]+df.schema.fields
    zipped_rdd=df.rdd.zipWithIndex()
        
    new_rdd= zipped_rdd.map(lambda x: ([x[1] + offset] + list(x[0])))
    )
    return spark.createDataFrame(new_rdd,new_schema)
```

```
%sql
create table keys(surrogate_id bigint comment 'surrogate key',id bigint comment 'identifier',
random_number double comment 'random number', str1 string comment 'string val1',str2 string comment 'string val2',str3 string comment 'string val3'
) using delta 
location '/FileStore/tables/sql_2/'
```

```
df=spark.sql("""
select id,rand() random_number, 'one' str1, 'two' str2, 'three' str3 from range(0, $(keys.nrows),1, $(keys.npartitions))""")
df_indexed= dfZipWithIndex(df, offset=1, colName='Surrogate_id')
df_indexed.createOrReplaceTempView('Change-set1')
spark.sql("""insert into keys select *from Change-set1""")
```
