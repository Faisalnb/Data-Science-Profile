```
from pyspark.sql import SparkSession
```

```
spark= SparkSession.builder.appName('Window').getOrCreate()
```

```
spark.read.csv('/FileStore/tables/df_tutors.csv',header=True).show()
```

```
tutors_df=spark.read.csv('/FileStore/tables/df_tutors.csv',header=True)
```

```
tutors_df.groupby('Department').agg({'Salary':'min'}).show()
```

```
tutors_df.groupby('Name').agg({'Salary':'avg'}).show()
```

```
from pyspark.sql.functions import cume_dist,lag,lead,avg,min,max
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

windowSpec=Window.partitionBy('Name').orderBy('Salary')
tutors_df.withColumn('avg',avg('Salary').over(windowSpec)).show()
```
