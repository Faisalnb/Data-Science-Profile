```
states= {"DL":"Delhi", "BL": "Bangalore" ,"MB": "Mumbai"}
```

```
broad_c= sc.broadcast(states)
data= [("emp1_name1","emp_surname1","IN","DL"),
       ("emp1_name2","emp_surname2","IN","BL"),
       ("emp1_name3","emp_surname3","IN","MB")]
```

```
rdd=sc.parallelize(data)
```

```
def is_state(key):
    return broad_c.value[key]
```

```
res= rdd.map(lambda x: (x[0],x[1],x[2],is_state(x[3]))).collect()
```

```
print(res)
```
