```
 iris_schema=StructType(
    [StructField('sepal_length',FloatType(),False), StructField('sepal_width',FloatType(),False), StructField('petal_length',FloatType(),False), StructField('petal_width',FloatType(),False), StructField('species',StringType(),True)])
dataframe1=spark.read.csv('/FileStore/tables/IRIS.csv',header=True,schema=iris_schema)
dataframe1.schema
```

```
def test(value):
    if value=='Iris-setosa':
        return 'X'
    else:
        return 'Y'
udf_test= f.udf(test,StringType())

df3= dataframe1.withColumn('der col',udf_test('Species'))
df3.show()
```

```
col= ['sr','name']
data=[('1','John'),('2','ron'),('3','amy')]
spark.createDataFrame(data,col).show()
```

```
df4=spark.createDataFrame(data,col)
```

```
def upperCase(str):
    return str.upper()
upperC_udf= udf(lambda x:upperCase (x), StringType() )
df4.withColumn('New name',upperC_udf('name'))\
.show(truncate=False)
```
