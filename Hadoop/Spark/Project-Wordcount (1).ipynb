```
from pyspark.sql.functions import trim,col,lower,split,explode,regexp_replace
```

```
def removePunctuation(column):
    return trim(lower(regexp_replace(column, "([^\s\w_]|_)+",""))).alias("sentence")
```

```
fileName ="/FileStore/shared_uploads/faizalnajeeb761@gmail.com/ebook.txt"
book_df= spark.read.text(fileName).select(removePunctuation(col('value')))
```

```
book_df.show(15)
```

```
book_split=book_df.select(split(book_df.sentence,"\s+").alias('split'))
```

```
display(book_split.show(5))
```

```
# after split do the explode to get a single col
book_clean=(book_split.select(explode(book_split.split).
                  alias('row')))
```

```
final_df=book_clean.where(book_clean.row.isNotNull())
```

```
final_df.count()
```

```
 def word_count(word_df):
        return (word_df.groupBy('row').count())
```
